```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("fifa22.csv")
```

```python
df.head()
```

Removing the longname column

```python
df=df.drop("long_name",axis=1)
df.head()
```

converting the "short name" column to lowercase.


```python
df["short_name"]=df["short_name"].apply(lambda x: x.lower())
df.head()
```

10 Most valuable players

```python
sorted_df=df.sort_values("value_eur",ascending=False)
valuable_players=sorted_df.iloc[0:10]
valuable_players
```

```python
plt.plot(df["overall"],df["potential"])
```

```python
#finding underrated players by analysing the potential/wage per week
df["potential/wage"]=df["potential"]/df["wage_eur"]
underrated_sorted_df=df.sort_values("potential/wage", ascending=False)
underrated_sorted_df.head()
```

```python
#finding underrated players by analysing the value/wage per week
df["value/wage"]=df["value_eur"]/df["wage_eur"]
underrated_sorted_df2=df.sort_values("value/wage",ascending=False)
underrated_sorted_df.head()
```

```python
df["height_m"]=df["height_cm"]/100
df.drop("height_cm",axis=1)
df.head()
```
