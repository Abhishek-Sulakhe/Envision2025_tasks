```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv("fifa22.csv")
df.head()
```

```python
columns=df[["overall","potential","value_eur", "wage_eur", "age", "height_cm"]]
columns.head()
columns.columns
```

```python
scaler=MinMaxScaler()
scaled_columns=scaler.fit_transform(columns)

scaled_columns=pd.DataFrame(scaled_columns)
scaled_columns
```

```python
scaled_columns.columns=['overall', 'potential', 'value_eur', 'wage_eur', 'age', 'height_cm']
scaled_columns
```

```python
df[['overall', 'potential', 'value_eur', 'wage_eur', 'age', 'height_cm']]=scaled_columns
```

```python
df.to_csv("scaled_processed.csv")
```
